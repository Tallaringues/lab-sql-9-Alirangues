How many distinct (different) actors' last names are there?

```python
data1 = pd.read_sql_query("SELECT count(DISTINCT last_name) AS 'Number of different last names' FROM actor", engine)
data1
```

Add an additional column day_type with values 'weekend' and 'workday' depending on the rental day of the week.

```python
query1 = "SELECT *, CASE WHEN ((WEEKDAY(rental_Date) >= 1) AND (WEEKDAY(rental_date) <= 5)) then 'workday' WHEN (WEEKDAY (rental_Date) = 6 OR WEEKDAY(rental_Date) = 7) then 'weekend'\
ELSE 'no_info' END AS day_type FROM rental"
```

```python
data2 = pd.read_sql_query(query1, engine)
display(data2['day_type'].unique())
display(data2.head())
```

Get all films with ARMAGEDDON in the title

```python
data3 = pd.read_sql_query("SELECT * FROM film WHERE title LIKE '%%ARMAGEDDON%%'", engine)
data3
```

Get 10 the longest films.

```python
data4 = pd.read_sql_query("SELECT * FROM film ORDER BY length DESC LIMIT 10", engine)
data4.head()
```

How many films include Behind the Scenes content?

```python
data5 = pd.read_sql_query("SELECT count(*) AS 'Num films behind scenes content' FROM film WHERE special_features LIKE '%%Behind the Scenes%%'", engine)
data5
```

Which kind of movies (rating) have a mean duration of more than two hours?

```python
data6 = pd.read_sql_query("SELECT rating, round(avg(length),2) AS mean_duration FROM film GROUP BY rating HAVING mean_duration>120", engine)
data6
```

Rank films by length (filter out the rows that have nulls or 0s in length column).
In your output, only select the columns title, length, and the rank.

```python
data7 = pd.read_sql_query("SELECT title, length, RANK() OVER (ORDER BY length) AS 'rank_duration' FROM film WHERE length != 0 AND length IS NOT NULL", engine)
data7

# ALTERNATIVE USING ROW_NUMBER()
#data7 = pd.read_sql_query("SELECT title, length, ROW_NUMBER() OVER (ORDER BY length) AS 'rank_duration' FROM film WHERE length != 0 AND length IS NOT NULL", engine)
#data7
```
